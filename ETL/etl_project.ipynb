```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
# Create birth df
births_file = "Resources/births.csv"
births_file_df = pd.read_csv(births_file)
births_file_df.head()
```

```python
# Create death df
deaths_file = "Resources/deaths.csv"
deaths_file_df = pd.read_csv(deaths_file)
deaths_file_df.head()
```

```python
# Create population df
population_file = "Resources/population.csv"
population_file_df = pd.read_csv(population_file)
population_file_df.head()
```

```python
# Create unemployment df
unemployment_file = "Resources/unemployment.csv"
unemployment_file_df = pd.read_csv(unemployment_file)
unemployment_file_df.head()
```

```python
# Rename birth columns
births_file_df = births_file_df.rename(columns={"Year":"year"
                                                ,"District Code" : "district_code"
                                                ,"District Name" : "district_name"
                                                ,"Neighborhood Code" : "neigbhor_code"
                                                ,"Neighborhood Name" : "neigbhor_name"
                                                ,"Gender" : "gender"
                                                ,"Number" : "number"})
births_file_df.head()
```

```python
# Rename death columns
deaths_file_df = deaths_file_df.rename(columns={"Year":"year"
                                                ,"District.Code" : "district_code"
                                                ,"District.Name" : "district_name"
                                                ,"Neighborhood.Code" : "neigbhor_code"
                                                ,"Neighborhood.Name" : "neigbhor_name"
                                                ,"Age" : "age"
                                                ,"Number" : "number"})
deaths_file_df.head()
```

```python
# Rename population columns
population_file_df = population_file_df.rename(columns={"Year":"year"
                                                ,"District.Code" : "district_code"
                                                ,"District.Name" : "district_name"
                                                ,"Neighborhood.Code" : "neigbhor_code"
                                                ,"Neighborhood.Name" : "neigbhor_name"
                                                ,"Gender" : "gender"
                                                ,"Age" : "age"
                                                ,"Number" : "number"})
population_file_df.head()
```

```python
# Remove columns from unemployment
del unemployment_file_df["Month"]
del unemployment_file_df["Gender"]
del unemployment_file_df["Demand_occupation"]
unemployment_file_df.head()
```

```python
# Rename unemployment columns
unemployment_file_df = unemployment_file_df.rename(columns={"Year":"year"
                                                ,"District Code" : "district_code"
                                                ,"District Name" : "district_name"
                                                ,"Neighborhood Code" : "neigbhor_code"
                                                ,"Neighborhood Name" : "neigbhor_name"
                                                ,"Number" : "number"})
unemployment_file_df.head()
```

```python
# Remove year 2013 and 2014 rows from births, population and unemployment
year_list = [2013, 2014]
births_file_df = births_file_df.loc[~births_file_df["year"].isin(year_list), :]
population_file_df = population_file_df.loc[~population_file_df["year"].isin(year_list), :]
unemployment_file_df = unemployment_file_df.loc[~unemployment_file_df["year"].isin(year_list), :]
```

```python
# Add age column in bith 
births_file_df["age"] = "0-4"
births_file_df.head()
```
